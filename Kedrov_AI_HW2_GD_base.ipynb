```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
import random
import time
import sys
```

```python
x = sp.Symbol('x')
```

```python
def f(x):
    return x**2 - 30*x + 1

def g(x):
    return -x**3 + 5*x - 7*x**2 + 17

def h(x):
    return x**2 - 1/x

def j(x):
    return 1/(x**5) + x**17 + 8
```

```python
def find_extremum_and_inflection_points(func, name):
    print(f"\n{name})")

    first_derivative = sp.diff(func(x), x)

    # точки экстремумы (приравниваем производную к 0)
    extremum_points = sp.solve(first_derivative, x)
    print(f"Точки экстремумов {name}: {extremum_points}")

    minimums = []
    maximums = []

    for id, point in enumerate(extremum_points):
        # подставляем точку в нашу функцию
        func_value = func(point)
        print(f"\t{id + 1}) {'(комплексное число)' if not point.is_real else ''} x = {point}, {name}({point}) = {func_value}")

        # берём вторую производную
        second_derivative = sp.diff(first_derivative, x)
        second_derivative_value = second_derivative.subs(x, point)

        # определяем тип экстремума
        if point.is_real:
          if second_derivative_value > 0:
              print(f"\t\tТочка x = {point} - минимум")
              minimums.append(point)
          elif second_derivative_value < 0:
              print(f"\t\tТочка x = {point} - максимум")
              maximums.append(point)
          else:
              print(f"\t\tТочка x = {point} - возможно перегиб")

    # точки перегиба (где вторая производная равна 0)
    inflection_points = sp.solve(second_derivative, x)
    print(f"Точки перегиба {name}: {inflection_points}")
    return {'first_derivative': first_derivative, 'extremum_points': extremum_points, 'second_derivative': second_derivative, 'minimums': minimums, 'maximums': maximums, 'inflection_points': inflection_points}

f_values = find_extremum_and_inflection_points(f, "f")
g_values = find_extremum_and_inflection_points(g, "g")
h_values = find_extremum_and_inflection_points(h, "h")
j_values = find_extremum_and_inflection_points(j, "j")
```

## Задание 6 (1 балл)

Модифицируйте функцию, осуществляющую метод градиентного спуска с семинара: добавьте опцию останавливать метод, если абсолютное значение производной функции в точке < некоторого $\delta$ (delta добавьте в аргументы функции).

```python
def gradient_descent(x_start, learning_rate, epsilon, delta, num_iterations, func):
    x_curr = x_start
    df_x = sp.diff(func(x))

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):
      grad_value = df_x.subs(x, x_curr)
      x_new = x_curr - learning_rate * df_x.subs(x, x_curr)
      trace.append(x_new)

      if abs(x_new - x_curr) < epsilon or abs(grad_value) < delta:
        return x_curr, trace

      x_curr = x_new

    return x_curr, trace
```

Подберите начальное приближение (сделайте несколько запусков из случайной начальной точки) и шаг градиентного спуска таким образом, чтобы процедура выдавала правильное значение наименьшего локального минимума функций (если он есть):

```python
# увеличение предела для обработки больших чисел
sys.set_int_max_str_digits(10000)


def gradient_descent(x_start, learning_rate, epsilon, delta, num_iterations, func):
    x_curr = x_start
    df_x = sp.diff(func(x))  # производная
    trace = [x_curr]

    for i in range(num_iterations):
        grad_value = df_x.subs(x, x_curr)

        # проверка градиента на переполнение
        if sp.Abs(grad_value) > 1e10:
            print(f"Переполнение градиента при x = {x_curr}, grad = {grad_value}")
            break

        # обновление значения
        x_new = x_curr - learning_rate * grad_value

        # ограничение на экстремальные значения x
        if abs(x_new) > 1e6:
            print(f"Значение x вышло за допустимые пределы: x = {x_new}")
            break

        trace.append(x_new)

        # условия останова
        if abs(x_new - x_curr) < epsilon or abs(grad_value) < delta:
            return x_new, func(x).subs(x, x_new), trace

        x_curr = x_new

    return x_curr, func(x).subs(x, x_curr), trace

def plot_function_and_minimum(func, x_min, y_min, trace, title, linspace_start, linspace_stop):
    x_vals = np.linspace(linspace_start, linspace_stop, 500)
    y_vals = [func(x).subs(x, val) for val in x_vals]

    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, label=f"{title}", color='blue')
    plt.scatter([x_min], [y_min], color='red', label='Minimum')

    # отрисовка точек градиентного спуска
    for point in trace:
        y_point = func(x).subs(x, point)
        plt.scatter([point], [y_point], color='green', s=10)

    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid()
    plt.show()

# гиперпараметры
learning_rate = 0.01
epsilon = 1e-6
delta = 1e-6
num_iterations = 1000

functions = {
    "f(x)": f,
    "g(x)": g,
    "h(x)": h,
    "j(x)": j
}

x0_values = {
    "f(x)": 0,
    "g(x)": 0,
    "h(x)": 1,
    "j(x)": 1
}

linspace_start_stop = {
    "f(x)": (-5, 20),
    "g(x)": (-7, 3),
    "h(x)": (-1.5, 1.5),
    "j(x)": (0.85, 1.025)
}

for name, func in functions.items():
    x0 = x0_values[name]
    linspace_start, linspace_stop = linspace_start_stop[name]

    try:
        x_min, y_min, trace = gradient_descent(x0, learning_rate, epsilon, delta, num_iterations, func)
        print(f"Функция: {name}")
        print(f"Стартовая точка: x0 = {x0}")
        print(f"Найденный минимум: x = {x_min}, f(x) = {y_min}")

        plot_function_and_minimum(func, x_min, y_min, trace, title=name,
                                  linspace_start=linspace_start, linspace_stop=linspace_stop)
    except Exception as e:
        print(f"Ошибка для функции {name}: {e}")
```

## Задание 8 (1.5 балла)

Реализуйте [метод Ньютона](https://en.wikipedia.org/wiki/Newton%27s_method_in_optimization) (метод второго порядка) для поиска минимума функции.

Про метод ещё можно почитать [здесь](https://math.semestr.ru/optim/minimum.php).

```python
# функция нахождения минимума методом Ньютона
def newton_min(func, x_start, epsilon=1e-6, max_iterations=1000):
  first_derivative = sp.diff(func(x), x)
  second_derivative = sp.diff(first_derivative, x)

  trace = [x_start]
  x_curr = x_start

  for iteration in range(max_iterations):
    grad_value = first_derivative.subs(x, x_curr)
    hessian_value = second_derivative.subs(x, x_curr)

    if hessian_value == 0:
      print("Ошибка: вторая производная равна нулю. Делить на ноль нельзя.")
      return None

    x_next = x_curr - grad_value / hessian_value

    # критерий останова по эпсилон
    if abs(x_next - x_curr) < epsilon:
      y_min = func(x_curr)
      return float(x_next), y_min, trace

    trace.append(x_curr)
    x_curr = x_next

  print(f"Метод Ньютона не сошёлся за {max_iterations} итераций.")
  return None
```

Проверьте, решает ли Momentum GD проблему поиска начального приближения для функций из задания 1?

```python
# градиентный спуск с импульсом
def momentum_gradient_descent(x_start, learning_rate, momentum, epsilon, num_iterations, func):
    x_curr = x_start
    df_x = sp.diff(func)
    trace = [x_curr]
    velocity = 0

    for i in range(num_iterations):
        grad_value = df_x.subs(x, x_curr)

        # обновление скорости и значения
        velocity = momentum * velocity - learning_rate * grad_value
        x_new = x_curr + velocity

        trace.append(x_new)

        # условия останова
        if abs(x_new - x_curr) < epsilon or abs(grad_value) < epsilon:
            return float(x_new), float(func.subs(x, x_new)), trace

        x_curr = x_new

    return float(x_curr), float(func.subs(x, x_curr)), trace

# определяем функции как SymPy выражения
f = x**2 - 30*x + 1
g = -x**3 + 5*x - 7*x**2 + 17
h = x**2 - 1/x
j = 1/(x**5) + x**17 + 8

# начальное приближение с momentum
def test_momentum_gd():
    functions = [f, g, h, j]
    learning_rate = 0.01
    momentum = 0.9
    epsilon = 1e-6
    num_iterations = 1000

    results = []

    for func in functions:
        for x_start in [-10, -1, 0.1, 1, 10]:
            momentum_result = momentum_gradient_descent(x_start, learning_rate, momentum, epsilon, num_iterations, func)
            results.append({
                'function': str(func),
                'x_start': x_start,
                'x_min': momentum_result[0],
                'y_min': momentum_result[1],
                'iterations': len(momentum_result[2])
            })

    for res in results:
        print(f"функция: {res['function']}, стартовая точка: {res['x_start']}, x_min: {res['x_min']}, y_min: {res['y_min']}, итераций: {res['iterations']}")

test_momentum_gd()
```

Momentum помогает найти минимум для функций при условии удачного начального приближения, но не гарантирует сходимость для функций с резкими изменениями градиента и множеством локальных минимумов.
